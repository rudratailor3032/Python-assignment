```python
n = int(input("enter the number:"))
rev = 0
while(n != 0):
    rem = n % 10
    rev = rev * 10 + rem 
    n = n // 10
print(f"reverse is number {rev}")
```

```python
n = int(input("enter the number: "))
rem = 0 
add = 0
while(n != 0):
    rem = n % 10
    add = add + rem
    n = n // 10
    print(f"sum of digits are : {add}" )
```
