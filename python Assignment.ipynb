```python
def is_perfect(n):
    r = 0
    for i in range(1,n):
        if n % i == 0:
            r += i 
    return r == n

print(is_perfect(6))
print(is_perfect(10))
print(is_perfect(13))
```

```python
def is_palindrome(r):
    i = 0 
    a = len(r) - 1

    while i < a:
        if r[i] != r[a]:
            return False
        i += 1
        a -= 1

    return True

print(is_palindrome("Rudra"))
print( is_palindrome("Saloni"))
```

```python
80) Write a Python program to count the frequency of words in a file.
```

```python
fs = open("file.txt","r")

text = fs.read().lower()
words = text.split()

freq={}

for i in words:
    if i in freq:
        freq[i] += 1
    else:
        freq[i] = 1
    
print(freq)    
```
